```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = np.array(cv.imread("train_image_000.png"))
grid = np.array(cv.imread("grid.png"))

grid[:, :, 1] = 0
grid[:, :, 2] = 0
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(gray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

contours = [c for c in contours if cv.contourArea(c) > 30 and len(c) > 2]
```

```python
img |= grid

search = np.zeros_like(img)
search |= grid

cv.drawContours(search, contours, -1, (0, 255, 0), 2)

cv.drawContours(img, contours, -1, (0, 255, 0), 2)

plt.imshow(search[3800:4400, 2300:2900])
# plt.imshow(img)
# plt.imshow(search)
plt.show()
```
