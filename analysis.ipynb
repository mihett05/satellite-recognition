```python
import cv2 as cv
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path

base_path = Path("vendor")
number = 11
img_path = str(base_path / "images" / f"train_image_0{number}.png")
mask_path = str(base_path / "masks" / f"train_mask_0{number}.png")
```

```python
mask = np.array(Image.open(mask_path))
mask = np.stack((mask * 255, np.zeros_like(mask), np.zeros_like(mask)), axis=2)

img = np.array(Image.open(img_path))

img_masked = mask | img

cv.imwrite("cum.png", cv.cvtColor(img_masked, cv.COLOR_RGB2BGR))

plt.imshow(img_masked)
plt.show()
```
